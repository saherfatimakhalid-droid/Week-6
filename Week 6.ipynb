```python
import numpy as np
import pandas as pd
from plotnine import *
```

```python
# load the dataset
data = pd.read_csv('diamonds.csv')
```

```python
# review the dataset
data.head()
```

```python
hist1 = (
    ggplot(data, aes(x="carat")) +
    geom_histogram(binwidth=0.2, fill="steelblue", color="black") +
    labs(
        title="Distribution of Diamond Carat",
        subtitle="Most diamonds have a carat less than 2",
        x="Carat",
        y="Count"
    ) +
    theme_minimal()
)

hist1.show() 
```

Most diamonds in the dataset are relatively small in size (carat < 2). Only a small proportion are very large diamonds, indicating skewed distribution.

```python
box1= (
    ggplot(data, aes(x="clarity", y="price", fill="clarity")) +
    geom_boxplot() +
    labs(
        title="Boxplot of Diamond Prices by Clarity",
        subtitle="Diamond price varies significantly with clarity",
        x="Clarity",
        y="Price (USD)"
    ) +
    theme_minimal()
)

box1.show() 
```

The boxplot shows that prices vary across clarity levels. Some clarity categories have much wider price ranges, while others are relatively stable. This suggests that clarity is a factor, but price is also influenced by other variables (like carat and cut).
